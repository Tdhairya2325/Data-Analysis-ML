```python
import matplotlib.pyplot as plt

x = [4, 5, 10, 4, 3, 11, 14 , 6, 10, 12, 4, 3, 7, 9, 10, 12, 11, 14, 13,  5, 14, 13, 2, 8, 6, 4, 1, 6, 3, 5, 6, 1, 3, 2, 5, 4, 1, 3]
y = [21, 19, 24, 17, 16, 25, 24, 22, 21, 21, 16, 23, 25, 17, 12, 11 ,14, 10, 9, 19, 10, 21, 20, 29, 18, 25, 17, 14, 12, 10, 11, 15, 14, 26, 25, 20, 29, 28]
print(len(x))
print(len(y))
plt.scatter(x, y)
plt.show()
```

```python
from sklearn.cluster import KMeans

data = list(zip(x, y))
print(data)
inertias = []

for i in range(1,34):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(data)
    inertias.append(kmeans.inertia_)

plt.plot(range(1,34), inertias, marker='o')
plt.title('Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
kmeans = KMeans(n_clusters=4)
kmeans.fit(data)

plt.scatter(x, y, c=kmeans.labels_)
plt.show()
```
